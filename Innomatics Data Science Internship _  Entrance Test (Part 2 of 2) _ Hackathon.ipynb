<h5>Q5. Which movie has recieved maximum number of user ratings</h5>
* Matrix<br>
* Pulp Fiction<br>
* Forrest Gump<br>
* Shawshank Redemption

```python
import pandas as pd

# Read the ratings.csv file
ratings = pd.read_csv('ratings.csv')

# Merge the ratings dataframe with the movies dataframe to get the movie titles
movies = pd.read_csv('movies.csv')
ratings_with_titles = pd.merge(ratings, movies, on='movieId')

# Group the ratings by movie title and count the number of ratings for each title
title_ratings = ratings_with_titles.groupby('title')['rating'].count()

# Find the movie title with the maximum number of ratings
max_ratings_title = title_ratings.idxmax()

# Print the movie title with the maximum number of ratings
print(max_ratings_title)
```

<h5>IMPORTANT NOTE 1</h5>
<b>Now that you have a good enough understanding of the given data, apply the "Mandatory Operation" given below before solving any of the following questions.<br>
Mandatory Operations:<br>
1. Group the user ratings based on movieId and apply aggregation operations like count and mean on ratings. <br>
2. Apply inner join on dataframe created from movies.csv and the grouped df from step 1.<br>
3. Filter only those movies which have more than 50 user ratings (i.e. > 50).<br>
    
Above steps will make sure that your data contains only those movies which has recieved more than 50 user ratings.</b>

```python
#1. Group the user ratings based on movieId and apply aggregation operations like count and mean on ratings.
grouped_ratings = ratings.groupby('movieId').agg(count=('rating', 'count'), mean=('rating', 'mean'))

# Flatten the column index of grouped_ratings
grouped_ratings.columns = grouped_ratings.columns.get_level_values(0)

#2. Apply inner join on dataframe created from movies.csv and the grouped df from step 1.
merged_df = movies.merge(grouped_ratings, on='movieId', how='inner')

# Filter only those movies which have more than 50 user ratings (i.e. > 50).
filtered_df = merged_df[merged_df['count'] > 50]
filtered_df.head()
```

</h5>

<h5>Q10. Select all the correct options which comes under top 5 popular movies based on number of user ratings</h5>

```python
#All the correct options which comes under top 5 popular movies based on number of user ratings

top_5_popular_movies = filtered_df.nlargest(5, 'count')['title']
top_5_popular_movies
```

<h5>Q11. Which Sci-Fi movie is "third most popular" based on the number of user ratings?</h5>

```python
# Filter the Dataset to include only Sci-Fi movies
sci_fi_movies = filtered_df[filtered_df['genres'].str.contains('Sci-Fi')]

# Sorting the Dataset by the count of ratings in descending order
sorted_sci_fi_movies = sci_fi_movies.sort_values('count', ascending=False)

# Select the movie at the third position
third_most_popular_sci_fi_movie = sorted_sci_fi_movies.iloc[2]['title']
third_most_popular_sci_fi_movie
```
